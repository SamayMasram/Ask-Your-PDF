```python
from dotenv import load_dotenv
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.documents import Document
from langchain_core.runnables import RunnablePassthrough
import os

load_dotenv()
# GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY") # This line is removed as it was overwriting the API key with None
```

```python
# 1. Load PDF using the path set in the previous cells
if '_pdf_path_for_loader' not in locals() or not _pdf_path_for_loader:
    raise ValueError(
        "PDF file path not set. Please run the upload cells first to set _pdf_path_for_loader."
    )

loader = PyPDFLoader(_pdf_path_for_loader)
docs = loader.load()
len(docs), docs[0].page_content[:500]
```

```python
# 2. Split into chunks
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
)
chunks = splitter.split_documents(docs)
len(chunks)
```

```python
# 3. FAISS vector store with Gemini embeddings
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001",
    api_key=GOOGLE_API_KEY
)

vectorstore = FAISS.from_documents(chunks, embeddings)
retriever = vectorstore.as_retriever(search_kwargs={"k": 4})

# 4. Prompt + Gemini chat model
prompt = ChatPromptTemplate.from_template(
    """
You are a helpful assistant. Use ONLY the context below to answer the question.
If the answer is not in the context, say you don't know.

Context:
{context}

Question:
{question}

Answer in a clear, concise way.
"""
)

def format_docs(docs: list[Document]) -> str:
    return "\n\n---\n\n".join(d.page_content for d in docs)

llm = ChatGoogleGenerativeAI(
    model="models/gemini-flash-latest",
    temperature=0.1,
    api_key=GOOGLE_API_KEY
)

rag_chain = (
    RunnablePassthrough.assign(context=(lambda x: x["question"]) | retriever | format_docs)
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
# Interactive Q&A loop with the LLM (using your existing rag_chain)

print("You can now ask questions about the PDF.")
print("Type 'exit' or 'quit' to stop.\n")

while True:
    user_question = input("Your question: ").strip()
    if user_question.lower() in {"exit", "quit"}:
        print("Goodbye!")
        break

    if not user_question:
        print("Please enter a non-empty question.\n")
        continue

    try:
        answer = rag_chain.invoke({"question": user_question})
        print("\nAnswer:")
        print(answer)
        print("\n" + "-" * 60 + "\n")
    except Exception as e:
        print(f"Error while generating answer: {e}")
        break
```

```python
# If needed (first time), install:
# !pip install -U langchain-community duckduckgo-search

from langchain_community.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.documents import Document
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_core.runnables import RunnableParallel, RunnablePassthrough

# assumes: chunks and GOOGLE_API_KEY already exist

# 1. PDF retriever
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001", # Updated to use an available model
    google_api_key=GOOGLE_API_KEY,
)
pdf_vectorstore = FAISS.from_documents(chunks, embeddings)
pdf_retriever = pdf_vectorstore.as_retriever(search_kwargs={"k": 4})

# 2. Web search tool
web_search = DuckDuckGoSearchRun()

# 3. Prompt
prompt = ChatPromptTemplate.from_template(
    """
You are a detailed assistant.
Use BOTH the PDF context and web search results to answer the question.

If there is any conflict:
- Prefer the PDF for details about the document.
- Use the web for general/world knowledge and recent information.

PDF context:
{pdf_context}

Web search context:
{web_context}

Question:
{question}

Provide a detailed, well-structured answer.
If something is unknown or uncertain, say so explicitly.
"""
)

def format_docs(docs: list[Document]) -> str:
    if not docs:
        return "No relevant PDF context found."
    return "\n\n---\n\n".join(d.page_content for d in docs)

# 4. LLM
llm = ChatGoogleGenerativeAI(
    model="models/gemini-flash-latest", # Updated to use an available model
    temperature=0.2,
    google_api_key=GOOGLE_API_KEY,
)

# 5. Combined chain (PDF + web)
pdf_web_chain = (
    RunnableParallel(
        question=RunnablePassthrough(),
        pdf_context=lambda q: format_docs(pdf_retriever.invoke(q)),
        web_context=lambda q: web_search.invoke(q),
    )
    | prompt
    | llm
    | StrOutputParser()
)
```
